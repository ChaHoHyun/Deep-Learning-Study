#####  Data

See, fork, and run a **random forest** benchmark model through Kaggle Scripts

- Period of bike rental data : 2 years
- train dataset : 19 days(1st ~ 19th) each month
- test dataset : about 10 days(20th ~ last) each month

We must predict the total count of bikes rented during 2 years for each month, hour.

- datetime - hourly date + timestamp (Year + Month + Day + Hour + Minutes + Seconds) 
- season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
- holiday - whether the day is considered a holiday
- workingday - whether the day is neither a weekend nor holiday
  (If it is not holiday or workingday, it will be **weekend**)
- weather 

  * 1: Clear, Few clouds, Partly cloudy, Partly cloudy
  
  * 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
  
  * 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
  
  * 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
  
- temp - temperature in Celsius
- atemp - "feels like" temperature in Celsius
- humidity - relative humidity
- windspeed - wind speed
- **casual** - number of non-registered user rentals initiated
- **registered** - number of registered user rentals initiated
- **count** - number of total rentals (casual + registered)

$$ casual + registered = count $$

- **Evaluation** : Submissions are evaluated one the Root Mean Squared Logarithmic Error (**RMSLE**). The RMSLE is calculated as

<br>

$$ RMSLE = \sqrt {\frac{1}{n}\sum\limits_{i=1}^n(log(p_i+1) - log(a_i + 1))^2} $$

### ML Modeling

```python
train.columns
```

```python
test.columns
```

```python
feature = ['season', 'holiday', 'workingday', 'weather', 'year_season', 
           'weekend', 'humidity','atemp', 'week_int', 'hour']
```

```python
x_train = train[feature]
y_train_casual = train['casual']
y_train_registered = train['registered']
y_train_count = train['count']
x_test = test[feature]
```

```python
figure, ((ax1), (ax2),(ax3)) = plt.subplots(nrows=3, ncols=1, figsize=(20,12))

train['casual'].plot(kind='kde', ax=ax1)
ax1.set_xlim([-10, 400])
train['registered'].plot(kind='kde', ax=ax2)
ax2.set_xlim([-10, 800])
train['count'].plot(kind='kde', ax=ax3)
ax3.set_xlim([-10, 1000]) 
```

```python
y_train_casual_log = np.log(y_train_casual + 1)
y_train_registered_log = np.log(y_train_registered + 1)
y_train_count_log = np.log(y_train_count + 1)

figure, ((ax1), (ax2),(ax3)) = plt.subplots(nrows=3, ncols=1, figsize=(20,12))

y_train_casual_log.plot(kind='kde', ax=ax1)
y_train_registered_log.plot(kind='kde', ax=ax2)
y_train_count_log.plot(kind='kde', ax=ax3)
```

```python
from sklearn.metrics import mean_squared_log_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import make_scorer

k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

model = RandomForestRegressor(n_estimators = 1000, n_jobs=-1)

y_predict_casual_log = cross_val_predict(model, x_train, y_train_casual_log, cv=k_fold, n_jobs=-1)
y_predict_registered_log = cross_val_predict(model, x_train, y_train_registered_log, cv=k_fold, n_jobs=-1)
y_predict_count_log = cross_val_predict(model, x_train, y_train_count_log, cv=k_fold, n_jobs=-1)

y_predict_casual = np.exp(y_predict_casual_log) - 1
y_predict_registered = np.exp(y_predict_registered_log) - 1
y_predict_count = np.exp(y_predict_count_log) - 1

y_predict = np.sqrt((y_predict_casual+y_predict_registered)*y_predict_count)

score = mean_squared_log_error(y_train_count, y_predict)
score = np.sqrt(score)

print("Score= {0:.5f}".format(score))
```

```python
model.fit(x_train, y_train_casual_log)
y_pred_casual_log = model.predict(x_test)
y_pred_casual = np.exp(y_pred_casual_log) - 1
y_pred_casual.shape
```

```python
model.fit(x_train, y_train_registered_log)
y_pred_registered_log = model.predict(x_test)
y_pred_registered = np.exp(y_pred_registered_log) - 1
y_pred_registered.shape
```

```python
sub_sample['count'] = y_pred_casual + y_pred_registered

sub_sample.to_csv("./data/Submission.csv", index=False)
```
