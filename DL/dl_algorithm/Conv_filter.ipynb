##### Mean value Blur

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

trump = cv2.imread('./model_img/image/trump.jfif')
image = cv2.cvtColor(trump, cv2.COLOR_BGR2RGB) # Convert image channel BGR to RGB
fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(image)
```

```python
blur_fliter = np.ones((12,12), dtype = np.float64) / 144
trump_blur = cv2.filter2D(image, -1, blur_fliter) # ddepth : Export Same image Size 
fig, ax = plt.subplots(1,2,figsize=(15,12))
ax[0].imshow(image)
ax[1].imshow(trump_blur)
```

##### Gaussian Blur

<br>

![image](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FcUECiH%2FbtqJUME9AwC%2FSWsLYpI5bmnUSgBNb5Fq1K%2Fimg.png)

```python
import cv2
import numpy as np

kernel1d = cv2.getGaussianKernel(12, 3) # Kernel size : 5, Sigma : 3
kernel2d = np.outer(kernel1d, kernel1d.transpose())
print(kernel2d.shape)
```

```python
trump_blur = cv2.filter2D(image, -1, kernel2d) # ddepth : Export Same image Size 
fig, ax = plt.subplots(1,2,figsize=(15,12))
ax[0].imshow(image)
ax[1].imshow(trump_blur)
```

#### Sharpening

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
trump_sharp = cv2.filter2D(image, -1, kernel)
fig, ax = plt.subplots(1,2,figsize=(15,12))
ax[0].imshow(image)
ax[1].imshow(trump_sharp)
```

#### Embossing
<br>
- embossing :  several techniques for creating a raised pattern on a material

```python
kernel = np.array([[-2, -1, 0],
                   [-1, 1, 1],
                   [0, 1, 2]])
trump_emb = cv2.filter2D(image, -1, kernel)
fig, ax = plt.subplots(1,2,figsize=(15,12))
ax[0].imshow(image)
ax[1].imshow(trump_emb)
```

#### Edge Detection

```python
kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
trump_edge = cv2.filter2D(image, -1, kernel)
fig, ax = plt.subplots(1,2,figsize=(15,12))
ax[0].imshow(image)
ax[1].imshow(trump_edge)
```
