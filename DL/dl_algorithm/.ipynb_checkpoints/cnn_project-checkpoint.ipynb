### Data Crawling by Selenium in Google

#### Shark image

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import time 
import urllib.request 
import os

search = "shark underwater"   # 이미지 이름
count = 5000    # 크롤링할 이미지 개수
saveurl = "./cnn_example/data/shark/"  # 이미지들을 저장할 폴더 주소

# 셀레니움으로 구글 이미지 접속 후 이미지 검색
options = webdriver.ChromeOptions()
options.headless = True
options.add_argument("window-size=1920x1080")

driver = webdriver.Chrome(options=options)  #options=options 
driver.get("https://www.google.co.kr/imghp?hl=ko&tab=wi&ogbl") 
elem = driver.find_element_by_name("q") 
elem.send_keys(search)
elem.send_keys(Keys.RETURN) 

# 페이지 끝까지 스크롤 내리기 
SCROLL_PAUSE_TIME = 1 
# 스크롤 깊이 측정하기 
last_height = driver.execute_script("return document.body.scrollHeight") 

# 스크롤 끝까지 내리기 
while True:  
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") 
    # 페이지 로딩 기다리기 
    time.sleep(SCROLL_PAUSE_TIME) 
    # 더 보기 요소 있을 경우 클릭하기 
    new_height = driver.execute_script("return document.body.scrollHeight") 
    if new_height == last_height: 
        try: 
            driver.find_element_by_css_selector(".mye4qd").click() 
        except: 
            break 
    last_height = new_height 

#이미지 찾고 다운받기
images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")
for i in range(count):
    try: 
        images[i].click() # 이미지 클릭
        time.sleep(1)

        imgUrl = driver.find_element_by_css_selector(".n3VNCb").get_attribute("src")
        urllib.request.urlretrieve(imgUrl, saveurl + str(i) + ".jpg")    # 이미지 다운
    except:
        pass
driver.close()
```

### Data Pre-processing

#### Load Sample Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

shark_example = cv2.imread('./cnn_example/data/shark/10.jpg')
fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(shark_example)
```

#### Load All image for train data

##### Load All image list by dictionary

```python
import os
import numpy as np
from PIL import Image

# Convert all images
folder_path = './cnn_example/data/'
folder_list = os.listdir(folder_path) # Find all directory in folder_path

def load_images_folder(folder_list):
    images_list = []
    for folder in folder_list:
        image_list = os.listdir(folder_path + folder)
        images = [img for img in image_list if img.endswith(".jpg")] # 지정된 확장자만 필터링
        images_list.append({"folder_name" : folder, 'image' : images})
    return images_list

images_list = load_images_folder(folder_list)
```

##### Convert all image file to Numpy array

```python
train_list = []
test_list = []
image_w, image_h = 64, 64

for i in range(len(folder_list)):
    count = 0
    for j in list(images_list[i].values())[1]:
        # Access each image by PIL library and Convert BW, GBR → RGB
        img = Image.open(folder_path + folder_list[i] + '/' + j).convert('RGB')
        # Unify all image size
        img = img.resize((image_w, image_h))
        # Convert image to numpy array 
        img_array = np.array(img)
        if img_array.shape[0] == 1:
            # Drop (1, 1) shape array
            pass
        else:
            if (img_array.ndim == 2):
                # Dimesion 2 (Black and White) → 3
                img_array = img_array.reshape(img_array.shape + (1,))
            train_list.append(img_array)
            test_list.append(folder_list[i])
            # Check Image's Label & Dimension & Shape 
#             print(f"Folder name : {folder_list[i]}, {i}, Dimension : {img_array.ndim}, Shape : {img_array.shape}")
            # Check counts of each Label
            count += 1
    print(count)
            
train, test = np.array(train_list), np.array(test_list)
```

##### Pre-processing for Train, test data

```python
size = []

for i in range(len(train)):
    size.append(train[i].shape[0])
    size.append(train[i].shape[1])
    
size_array = np.array(size)
min(size_array)
```

```python
train[0].shape
```

```python
# Count all data by dividing Label (Shark, Dolphin, Whale)
np.unique(test, return_counts=True)[1]
```

```python
# Label Encoding for Test dataset
test = np.unique(test, return_inverse=True)[1]
```

##### Train & test Split

```python
len(train) , len(test)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(train, test,
                                                   test_size = 1/ 5,
                                                   random_state = 0)
```

```python
print(len(y_train) + len(y_test))
len(x_train) , len(x_test), len(y_train) , len(y_test)
```

##### Normalization

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print(f'Input shape : {x_train[0].shape}')
```

### Modeling

#### CNN Model (General)

```python
from tensorflow.keras import Input, Model, optimizers
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import plot_model

batch_size = 10000 # 2048 -> 10,000: 너무 오래걸림
num_classes = len(np.unique(y_train))
epochs = 3
input_shape = x_train[0].shape
```

```python
inputs = Input(shape=input_shape)
x = Conv2D(filters= 32, kernel_size= (3,3),
          activation = 'relu')(inputs)
x = Conv2D(filters= 64, kernel_size= (3,3),
          activation = 'relu')(x)
x = MaxPooling2D(pool_size=(2,2))(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
outputs = Dense(num_classes, activation='softmax')(x)

model = Model(inputs = inputs, outputs = outputs)

model.compile(loss='sparse_categorical_crossentropy',
              optimizer=optimizers.Adam(),
              metrics=['accuracy'])

plot_model(model, show_shapes = True)
```

```python
model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1)
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

#### Hyper Parameter Tunning

```python
train_list = []
test_list = []
image_w = 224

for i in range(len(folder_list)):
    count = 0
    for j in list(images_list[i].values())[1]:
        img = Image.open(folder_path + folder_list[i] + '/' + j).convert('RGB')
        img = img.resize((image_w, image_w))
        img_array = np.array(img)
        if img_array.shape[0] == 1:
            pass
        else:
            if (img_array.ndim == 2):
                img_array = img_array.reshape(img_array.shape + (1,))
            train_list.append(img_array)
            test_list.append(folder_list[i])
            count += 1
    print(count)
            
train, test = np.array(train_list), np.array(test_list)
```

```python
size = []

for i in range(len(train)):
    size.append(train[i].shape[0])
    size.append(train[i].shape[1])
    
size_array = np.array(size)
min(size_array)
```

```python
# Label Encoding for Test dataset
test = np.unique(test, return_inverse=True)[1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(train, test,
                                                   test_size = 1/ 5,
                                                   random_state = 0)

print(len(y_train) + len(y_test))
len(x_train) , len(x_test), len(y_train) , len(y_test)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print(f'Input shape : {x_train[0].shape}')
```

```python
from tensorflow.keras import Input, Model, optimizers
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import plot_model

batch_size = 1000 # 2048 -> 10,000: 너무 오래걸림
num_classes = len(np.unique(y_train))
epochs = 20
input_shape = x_train[0].shape
```

```python
inputs = Input(shape=input_shape)
x = Conv2D(filters= 32, kernel_size= (3,3),
          activation = 'relu')(inputs)
x = Conv2D(filters= 64, kernel_size= (3,3),
          activation = 'relu')(x)
x = MaxPooling2D(pool_size=(2,2))(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
outputs = Dense(num_classes, activation='softmax')(x)

model = Model(inputs = inputs, outputs = outputs)

model.compile(loss='sparse_categorical_crossentropy',
              optimizer=optimizers.Adam(),
              metrics=['accuracy'])

plot_model(model, show_shapes = True)
```

```python
history = model.fit(x_train, y_train,
                  batch_size=batch_size,
                  epochs=epochs,
                  verbose=1,
                   validation_data = (x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
plt.figure(figsize=(12, 10))
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss")
```
