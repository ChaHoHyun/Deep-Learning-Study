#### Deep Neural Network

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras import Model, Input, layers
from tensorflow.keras import losses, optimizers
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
# plt.rc("figure", figsize=(12,10))
plt.figure(figsize=(8,6))

batch_size = 4000
num_classes = 10
epochs = 20

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

plt.figure()
plt.imshow(x_train[0], cmap='gray')
plt.show()

print(x_train[0].shape, y_train[0].shape)
print(y_train[0])
```

```python
print(x_train.shape)
x_train.reshape(60000, x_train.shape[1]*x_train.shape[2]).shape
```

```python
x_train = x_train.reshape(60000, x_train.shape[1]*x_train.shape[2])
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
# 이 위까지는 픽셀별 값이 0~255
x_train /= 255
x_test /= 255
# 이제 표준화를 해서 픽셀별 값을 0~1
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
import tensorflow.keras

# convert class vectors to binary class matrices
# Don't need to convert 'to_categorical' by using loss function "sparse sparse_categorical_crossentropy"
# y_train = tensorflow.keras.utils.to_categorical(y_train, num_classes)
# y_test = tensorflow.keras.utils.to_categorical(y_test, num_classes)

# model generation
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import utils

# functionnal API
inputs = Input(x_train[0].shape)
x = layers.Dense(1024, activation = 'relu')(inputs)
x = layers.Dense(1024, activation = 'relu')(x)
x = layers.Dense(512, activation = 'relu')(x)
x = layers.Dense(32, activation = 'relu')(x)
outputs = layers.Dense(10, activation = 'softmax')(x)

model = Model(inputs, outputs)

dot_img_file = './model_img/mnist/mnist_dnn.png'
utils.plot_model(model, show_shapes = True, to_file=dot_img_file)
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs = epochs,
                    verbose = 1,
                    validation_data = (x_test, y_test))
```

```python
score = model.evaluate(x_test, y_test, verbose = 1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
plt.figure(figsize=(12, 10))
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss")
```

#### Convolution Neural Network

##### Sequential API

```python
import tensorflow.keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras import backend as K

batch_size = 10000 # 2048 -> 10,000: 너무 오래걸림
num_classes = 10
epochs = 3

# input image dimensions
img_rows, img_cols = 28, 28

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# data format : 28*28*1 or 1*28*28 채널이 앞이냐 뒤냐? => tensorflow는 channel이 맨 뒤! 무시해도 되는 code임
if K.image_data_format() == 'channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

# 0~1 사이의 normalize
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

print(input_shape)
# convert class vectors to binary class matrices
#y_train = tensorflow.keras.utils.to_categorical(y_train, num_classes)
#y_test = tensorflow.keras.utils.to_categorical(y_test, num_classes)

# build your model

model = Sequential()
model.add(Conv2D(filters = 32, kernel_size = (3,3), #strides=1,  shift 1칸씩 이동
                activation = 'relu', input_shape = input_shape))
model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation = 'softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer=tensorflow.keras.optimizers.Adam(),
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1)
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

model.save('./model_img/mnist/my_cnn_model.h5')
del model
```

##### Funtional API

```python
from tensorflow.keras.utils import plot_model

inputs = Input(shape=input_shape)
x = Conv2D(filters= 32, kernel_size= (3,3),
          activation = 'relu')(inputs)
x = Conv2D(filters= 64, kernel_size= (3,3),
          activation = 'relu')(x)
x = MaxPooling2D(pool_size=(2,2))(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
outputs = Dense(num_classes, activation='softmax')(x)

model = Model(inputs = inputs, outputs = outputs)

model.compile(loss='sparse_categorical_crossentropy',
              optimizer=tensorflow.keras.optimizers.Adam(),
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1)
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
plot_model(model, show_shapes = True)
```

### Cifar-10 Dataset

<br>

![image](https://production-media.paperswithcode.com/datasets/4fdf2b82-2bc3-4f97-ba51-400322b228b1.png)

<br>

The CIFAR-10 dataset consists of 60000 32x32 colour images in 10 classes, with 6000 images per class. There are 50000 training images and 10000 test images.

```python
import tensorflow.keras
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D

batch_size = 32 # 2의 배수로 설정! : * batch는 64 128 256 등 2의 거듭제곱로 사용하는 이유는 GPU의 메모리가 2의 거듭제곱이기 때문입니다. 따라서, 최고의 효율을 내기 위해 2의 거듭제곱으로 설정합니다. 
#                                     일반적으로 통용되는 사실은, Batch size가 작을수록, 시간은 오래 걸리지만 효과는 좋다는 점입니다
num_classes = 10
epochs = 5
num_predictions = 20

# The data, split between train and test sets:
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

print(y_test)
```

```python
import matplotlib.pyplot as plt
i = 2
plt.imshow(x_test[i])
print(y_test[i])
```

```python
input_shape = x_train[0].shape
print(input_shape)
input_shape = x_train.shape[1:]
print(input_shape)
```

```python
from tensorflow.keras import Input, Model
from tensorflow.keras.utils import plot_model

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255 # image는 0~255 값이므로 0부터 1까지의 값으로 normalization! => 모델에서 더 좋은 성능을 보여줌!
x_test /= 255

# input_shape = (32, 32, 3)

input_layer = Input(shape=x_train.shape[1:])
x = Conv2D(filters= 32, kernel_size= (3,3), activation = 'relu', padding='same')(input_layer)
x = Conv2D(filters= 32, kernel_size= (3,3), activation = 'relu', padding='same')(x)
x = MaxPooling2D(pool_size=(2,2))(x)
x = Dropout(0.25)(x)

x = Conv2D(filters= 32, kernel_size= (3,3), activation = 'relu', padding='same')(x)
x = MaxPooling2D(pool_size=(2,2))(x)
x = Dropout(0.25)(x)

x = Flatten()(x)

x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(num_classes, activation='softmax')(x)

model = Model(inputs = input_layer, outputs = output)

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(x_train, y_train,
         batch_size=batch_size,
         epochs=epochs,
         validation_data=(x_test, y_test))

# Score trained model.
scores = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])

model.summary()
```

### VGG16 with Webcam

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions

from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
import numpy as np

model = VGG16(weights='imagenet')

model.summary()

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    

    # Display the resulting frame
    
    dst = cv2.resize(frame, dsize=(224, 224), interpolation=cv2.INTER_AREA)    
    img = img_to_array(dst)
    img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))
    img = preprocess_input(img)
    yhat = model.predict(img)

    label = decode_predictions(yhat)
    label = label[0][0]
    
    cv2.imshow('frame',dst)
    
    if cv2.waitKey(1) & 0xFF == ord('w'):
        print('%s (%.2f%%)' % (label[1], label[2]*100))    
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
